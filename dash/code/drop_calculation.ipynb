#### 計算 dipping 程度，公式：(awake_var - sleep_var) / (awake_var)
Calculate the proportion of blood pressure and heart rate drop, formula (awake_var - sleep_var) / (awake_var)
var = variables

```python
import pandas as pd
import numpy as np
import os
```

```python
dir_path = 'C:/Users/Yao-Chung Chen/OneDrive/github_ferygood/CR-Hemodynamics/dash/code/Clean_data/'
file = os.path.join(dir_path, 'control_and_dash.csv')
df = pd.read_csv(file, index_col=0)
df.head()
```

寫一個函式來計算 dipping 並加上表格: calculate_dipping()
write a function to calculate the drop proportioin and add the result to a new column

```python
def calculate_dipping(df):
    '''
    formula = (awake_var - sleep_var) / (awake_var)
    '''
    df['drop_sbp'] = (df['mean_awake_sbp'] - df['mean_sleep_sbp']) / df['mean_awake_sbp']
    df['drop_dbp'] = (df['mean_awake_dbp'] - df['mean_sleep_dbp']) / df['mean_awake_dbp']
    df['drop_rate'] = (df['mean_awake_rate'] - df['mean_sleep_rate']) / df['mean_awake_rate']
    return df
```

```python
df_dip = calculate_dipping(df)
df_dip.head()
```

```python
# save the file as control_and_dash_dip.csv
df_dip.to_csv(dir_path + 'control_and_dash_dip.csv')
```
